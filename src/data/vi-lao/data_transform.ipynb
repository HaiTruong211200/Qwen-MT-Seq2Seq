```python
import csv
import json

input_csv = "val_combined.csv"        # file CSV đầu vào
output_jsonl = "valid.vi-lo.general_trans.wmt23.jsonl"  # file JSONL đầu ra

with open(input_csv, "r", encoding="utf-8") as csvfile, \
     open(output_jsonl, "w", encoding="utf-8") as jsonlfile:
    
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        lao_sentence = row["lo"].strip()
        vi_sentence = row["vi"].strip()
        
        data = {
            "src_lang": "lo",          # Lào
            "tgt_lang": "vi",          # Việt
            "translation": {
                "lo": lao_sentence,
                "vi": vi_sentence
            },
            "data_name": "VOVNews",
            "task_type": "general_trans"
        }
        
        jsonlfile.write(json.dumps(data, ensure_ascii=False) + "\n")

print("Done!")
```

```python
import csv
import random

input_csv = "vi_lo_0k_20k.csv"

train_csv = "train.csv"
val_csv = "val.csv"
test_csv = "test.csv"

VAL_SIZE = 500
TEST_SIZE = 500

random.seed(42)  # reproducible

# Đọc toàn bộ dữ liệu
with open(input_csv, "r", encoding="utf-8") as f:
    rows = list(csv.DictReader(f))

total = len(rows)

if total < VAL_SIZE + TEST_SIZE:
    raise ValueError("Dataset không đủ số lượng câu!")

# Shuffle
random.shuffle(rows)

# Chia tập
val_rows = rows[:VAL_SIZE]
test_rows = rows[VAL_SIZE:VAL_SIZE + TEST_SIZE]
train_rows = rows[VAL_SIZE + TEST_SIZE:]


def write_csv(filename, data_rows):
    with open(filename, "w", encoding="utf-8", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=["vi", "lo"])
        writer.writeheader()
        
        for row in data_rows:
            writer.writerow({
                "vi": row["output"].strip(),   # tiếng Việt
                "lo": row["input"].strip()   # tiếng Lào
            })


write_csv(train_csv, train_rows)
write_csv(val_csv, val_rows)
write_csv(test_csv, test_rows)

print("Done!")
print(f"Train: {len(train_rows)}")
print(f"Val: {len(val_rows)}")
print(f"Test: {len(test_rows)}")
```
